```python
import numpy as np
import pandas as pd
import seaborn as sns

from pandas.plotting import scatter_matrix
```

```python
df=pd.read_csv('test.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(data = df, x = "Age", y = "Fare")

plt.show()
```

```python
titanic_pcalss_fig, titanic_pcalss_ax = plt.subplots()
color = ['blue', 'red', 'skyblue']
count = 0
for name, group in df.groupby('Pclass'):
    titanic_pcalss_ax.plot(group.Age, group.Fare, '.', label = name, alpha = 0.6, c = color[count])
    count += 1

titanic_pcalss_ax.legend(numpoints=1, title = "Passenger class", fontsize = 10)

plt.xlabel('Age')
plt.ylabel('Fare')
titanic_pcalss_ax.set_xlim(-1, 85)
titanic_pcalss_ax.set_ylim(-1, 600)

plt.show(titanic_pcalss_fig)
```

```python
sns.lineplot(data=df, x="Age", y="Fare")
```

```python
titanic_hist = df.Fare.plot.hist(bins = 40, color = 'skyblue')
plt.xlabel('Fare')

plt.show(titanic_hist)
```

```python
pclass_fare_titanic =  df[['Pclass', 'Fare']].pivot(columns = 'Pclass', values = 'Fare')

box_color = dict(boxes = 'red', whiskers = 'red', medians = 'blue', caps = 'red')

titanic_pclass_boxplot = pclass_fare_titanic.plot.box(color = box_color)
plt.xlabel('Passenger class')
plt.ylabel('Fare')

plt.show(titanic_pclass_boxplot)
```

```python
contingency_titanic = df.groupby(['Pclass', 'Parch']).size().unstack()
titanic_barplot = contingency_titanic.plot.bar(stacked=True, 
                                               color = ["skyblue", "blue"])
plt.ylabel("Count")
plt.xlabel('Passenger class')
plt.xticks(rotation=0)
plt.show(titanic_barplot)
```
